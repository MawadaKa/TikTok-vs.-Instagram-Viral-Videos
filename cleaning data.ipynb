```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import ast
pd.set_option('display.max_columns', None)
```

```python
tiktok = pd.read_csv('tiktok_dataset.csv')
ig = pd.read_csv('insta_reels.csv')
```

```python
#I'll drop the unwanted columns:
columns_to_drop_tiktok = ['#', 'claim_status', 'author_ban_status', 'video_id' ]
tiktok = tiktok.drop(columns=columns_to_drop_tiktok)
```

```python
ig.head()
```

```python
#I'll drop the unwanted columns:
columns_to_drop_ig = ['inputUrl', 'id', 'type', 'shortCode', 'url', 'firstComment', 'latestComments', 'dimensionsHeight', 'dimensionsWidth', 'displayUrl', 'images', 'videoUrl', 'alt', 'childPosts', 'ownerFullName', 'ownerUsername', 'ownerId', 'productType', 'isSponsored', 'taggedUsers', 'musicInfo', 'coauthorProducers', 'locationName', 'locationId', 'isPinned', 'paidPartnership', 'sponsors', 'error', 'description' ]
ig = ig.drop(columns=columns_to_drop_ig)
```

```python
ig.head()
```

```python
tiktok.isnull().sum()
```

```python
#since there's only 298 null values in multiple columns, I'll drop them
tiktok.dropna(subset=['video_view_count'], inplace=True)
tiktok.reset_index(drop=True, inplace=True)
```

```python
ig.isnull().sum()
```

```python
ig.dropna(subset=['hashtags'], inplace=True)
ig.reset_index(drop=True, inplace=True)
```

```python
#since there's 152 null values in videoPlayCount, I'll fill them with the mean value
mean_play = ig['videoPlayCount'].mean() 
ig['videoPlayCount'].fillna(mean_play, inplace=True)
```

```python
#the null values in the caption indicates that there's no caption added, so I'll fill them with 0
ig['caption'] = ig['caption'].fillna(0)
```

```python
#I want to encode the 'verified_status' column to 0 and 1
display(tiktok['verified_status'].value_counts(dropna=False))
tiktok['verified_status'] = tiktok['verified_status'].map({'verified': 1, 'not verified': 0})
```

```python
#I want to replace the values in 'hashtags' to be the number of hashtags
# Convert string representations of lists to actual lists
ig['hashtags'] = ig['hashtags'].apply(ast.literal_eval)

# Replace list values with the number of items in each list
ig['hashtags'] = ig['hashtags'].apply(lambda x: len(x) if isinstance(x, list) else 0)
```

```python
#I want to replace the values in 'mentions' to be the number of mentions
# Convert string representations of lists to actual lists
ig['mentions'] = ig['mentions'].apply(ast.literal_eval)

# Replace list values with the number of items in each list
ig['mentions'] = ig['mentions'].apply(lambda x: len(x) if isinstance(x, list) else 0)
```

```python
# Convert 'Timestamp' to datetime
ig['timestamp'] = pd.to_datetime(ig['timestamp'])

# Extract day of the week
ig['day'] = ig['timestamp'].dt.day_name()

# Extract date and time (hour and minute)
ig['time'] = ig['timestamp'].dt.strftime('%H:%M')

# Drop the original 'Timestamp' column
ig = ig.drop(columns=['timestamp'])
```

```python
ig['time'] = pd.to_datetime(ig['time'], format='%H:%M')

# Extract hour and minute
ig['hour'] = ig['time'].dt.hour

# Drop the original 'time' column
ig = ig.drop(columns=['time'])
```

```python
d_tiktok = tiktok.duplicated()
display(d_tiktok.value_counts(dropna=False))
#we have no duplicates in the tiktok dataframe
```

```python
d_ig = ig.duplicated()
display(d_ig.value_counts(dropna=False))
#we have no duplicates in the ig dataframe
```

```python
#unifying columns names
tiktok.rename(columns={'video_duration_sec': 'duration', 'video_transcription_text': 'caption', 'video_view_count': 'views', 'video_like_count': 'likes', 'video_share_count': 'shares', 'video_download_count': 'downloads', 'video_comment_count': 'comments'}, inplace=True)
ig.rename(columns={'commentsCount': 'comments', 'likesCount': 'likes', 'videoViewCount': 'views', 'videoPlayCount': 'plays', 'videoDuration': 'duration'}, inplace=True)
```

```python
new_order_tiktok = ['views', 'duration', 'caption', 'verified_status', 'likes', 'comments', 'shares', 'downloads']

# Reorder the columns
tiktok = tiktok[new_order_tiktok]

new_order_ig = ['plays', 'duration', 'caption', 'mentions', 'hashtags', 'likes', 'comments', 'views', 'hour', 'day']

# Reorder the columns
ig = ig[new_order_ig]
```
